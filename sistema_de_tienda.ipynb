```python

"Sistema de Tienda TechStore (versión seminteractiva para Colab)"


"--- Datos de ejemplo ---"
productos = [
    {'id': 1, 'nombre': 'iPhone 15', 'marca': 'Apple', 'categoria': 'Smartphone', 'precio': 999.99, 'stock': 10, 'disponible': True},
    {'id': 2, 'nombre': 'Samsung Galaxy S24', 'marca': 'Samsung', 'categoria': 'Smartphone', 'precio': 899.99, 'stock': 8, 'disponible': True},
    {'id': 3, 'nombre': 'MacBook Air M3', 'marca': 'Apple', 'categoria': 'Laptop', 'precio': 1299.99, 'stock': 5, 'disponible': True},
    {'id': 4, 'nombre': 'Dell XPS 13', 'marca': 'Dell', 'categoria': 'Laptop', 'precio': 1199.99, 'stock': 0, 'disponible': False},
    {'id': 5, 'nombre': 'Sony WH-1000XM5', 'marca': 'Sony', 'categoria': 'Audífonos', 'precio': 399.99, 'stock': 15, 'disponible': True}
]

empleados = [
    {'id': 101, 'nombre': 'Ana', 'apellido': 'García', 'departamento': 'Ventas', 'salario': 35000, 'activo': True},
    {'id': 102, 'nombre': 'Carlos', 'apellido': 'López', 'departamento': 'Técnico', 'salario': 42000, 'activo': True},
    {'id': 103, 'nombre': 'María', 'apellido': 'Rodríguez', 'departamento': 'Ventas', 'salario': 38000, 'activo': False},
    {'id': 104, 'nombre': 'José', 'apellido': 'Martínez', 'departamento': 'Inventario', 'salario': 30000, 'activo': True}
]

" --- Funciones ---"
def buscar_producto_por_nombre(productos, nombre):
    return [p for p in productos if nombre.lower() in p['nombre'].lower()]

def buscar_producto_por_id(productos, id_producto):
    for p in productos:
        if p['id'] == id_producto:
            return p
    return None

def buscar_producto_por_categoria(productos, categoria):
    return [p for p in productos if categoria.lower() == p['categoria'].lower()]

def buscar_productos_disponibles(productos):
    return [p for p in productos if p['stock'] > 0]

def buscar_producto_por_marca(productos, marca):
    return [p for p in productos if marca.lower() == p['marca'].lower()]

def buscar_empleado_por_nombre(empleados, nombre, apellido):
    for e in empleados:
        if e['nombre'].lower() == nombre.lower() and e['apellido'].lower() == apellido.lower():
            return e
    return None

def buscar_empleado_por_departamento(empleados, departamento):
    return [e for e in empleados if departamento.lower() == e['departamento'].lower()]

def buscar_empleados_activos(empleados):
    return [e for e in empleados if e['activo']]

" --- Ejecución simple (sin bucle) ---"
print("=== SISTEMA TECHSTORE ===")
print("Opciones disponibles:")
print("1. Buscar producto por nombre")
print("2. Buscar producto por ID")
print("3. Buscar producto por categoría")
print("4. Buscar productos disponibles")
print("5. Buscar producto por marca")
print("6. Buscar empleado por nombre completo")
print("7. Buscar empleado por departamento")
print("8. Ver empleados activos")

opcion = input("Seleccione una opción (1-8): ")

if opcion == "1":
    nombre = input("Ingrese el nombre del producto: ")
    print(buscar_producto_por_nombre(productos, nombre))
elif opcion == "2":
    pid = int(input("Ingrese el ID: "))
    print(buscar_producto_por_id(productos, pid))
elif opcion == "3":
    categoria = input("Ingrese la categoría: ")
    print(buscar_producto_por_categoria(productos, categoria))
elif opcion == "4":
    print(buscar_productos_disponibles(productos))
elif opcion == "5":
    marca = input("Ingrese la marca: ")
    print(buscar_producto_por_marca(productos, marca))
elif opcion == "6":
    nombre = input("Nombre del empleado: ")
    apellido = input("Apellido del empleado: ")
    print(buscar_empleado_por_nombre(empleados, nombre, apellido))
elif opcion == "7":
    dep = input("Departamento: ")
    print(buscar_empleado_por_departamento(empleados, dep))
elif opcion == "8":
    print(buscar_empleados_activos(empleados))
else:
    print(" Opción inválida.")
```
